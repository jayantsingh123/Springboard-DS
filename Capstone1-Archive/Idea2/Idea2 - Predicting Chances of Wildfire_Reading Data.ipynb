#### The original dataset from Kaggle (available at: https://www.kaggle.com/rtatman/188-million-us-wildfires) is in a SQLite format and the dates are in Julian date format. The dataset is converted into a pandas dataframe by using sqlite3 and connecting to the dataframe. Only the following columns of interest are used in creating the dataframe  -  Fire Year, Cause of Fire, Latitude, Longitude, State, Fips Code, Fips Name, Discovery Date, Discovery Time, Contained Date, Contained Time and Fire size. The dataframe is written into a table called Fires. 

```python
cnx = sqlite3.connect('/Users/shridhar/Downloads/FPA_FOD_20170508.sqlite')
```

```python
df = pd.read_sql_query("SELECT FIRE_YEAR,STAT_CAUSE_DESCR,LATITUDE,LONGITUDE,STATE,FIPS_CODE,FIPS_NAME,DISCOVERY_DATE,DISCOVERY_TIME,CONT_DATE,CONT_TIME,FIRE_SIZE FROM 'Fires'", cnx)
```

#### The dataframe contains information from many states, but for the purpose of this project, I'm only looking at California. Creating a csv file as Fires_CA with data for California. 

```python
df_CA = df[(df['STATE'] == 'CA') & (df['FIRE_YEAR']>=2010)]
```
